### Heroes Of Pymoli Data Analysis

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(20)
```

## Player Count

* Display the total number of players


```python
players = len(purchase_data["SN"].unique())
total_players = pd.DataFrame({"Total Players": [players]})
total_players 
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data["Item ID"].unique())
avg_price = purchase_data["Price"].mean()
num_purch = purchase_data["Purchase ID"].count()
revenue = purchase_data["Price"].sum()
tot_purch_analysis = pd.DataFrame({"Number of Unique Items": [unique_items],
                               "Average Price": [avg_price],
                               "Number of Purchases": [num_purch],
                               "Total Revenue": [revenue]})

tot_purch_analysis ["Average Price"] = tot_purch_analysis ["Average Price"].map("${:.2f}".format)
tot_purch_analysis ["Total Revenue"] = tot_purch_analysis ["Total Revenue"].map("${:.2f}".format)
tot_purch_analysis = tot_purch_analysis[["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]

tot_purch_analysis 
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
drop_dup = purchase_data.drop_duplicates(subset='SN', keep="first")

tot_gen_df = drop_dup["Gender"].count()

male_gen_df = drop_dup["Gender"].value_counts()['Male']

fem_gen_df = drop_dup["Gender"].value_counts()['Female']

non_dis_df = tot_gen_df - male_gen_df - fem_gen_df

percent_male_df = (male_gen_df / tot_gen_df) * 100
percent_fem_df = (fem_gen_df / tot_gen_df) * 100
percent_non_df = (non_dis_df / tot_gen_df) * 100


gender_demographics_df = pd.DataFrame({"": ['Male', 'Female', 'Other/Non-Disclosed'],
							"Total Count": [male_gen_df, fem_gen_df, non_dis_df],
                            "Percentage of Players": [percent_male_df, percent_fem_df, percent_non_df]})


gender_demographics_df ["Percentage of Players"] = gender_demographics_df ["Percentage of Players"].map("{:.2f}%".format)
gender_demographics_df  = gender_demographics_df .set_index('')
gender_demographics_df 
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gen_grp_df = purchase_data.groupby(["Gender"])

purch_cnt = gen_grp_df["Purchase ID"].count()
purch_price = gen_grp_df["Price"].mean()
tot_purch_value = gen_grp_df["Price"].sum()

drop_dup = purchase_data.drop_duplicates(subset='SN', keep="first")
grp_dup = drop_dup.groupby(["Gender"])

avg_tot_purch_pp = (gen_grp_df["Price"].sum() / grp_dup["SN"].count())

gen_purch_analysis = pd.DataFrame({"Purchase Count": purch_cnt,
                              "Average Purchase Price": purch_price,
                              "Total Purchase Value": tot_purch_value,
                              "Avg Total Purchase per Person": avg_tot_purch_pp})

gen_purch_analysis["Average Purchase Price"] = gen_purch_analysis["Average Purchase Price"].map("${:.2f}".format)
gen_purch_analysis["Total Purchase Value"] = gen_purch_analysis["Total Purchase Value"].map("${:.2f}".format)
gen_purch_analysis["Avg Total Purchase per Person"] = gen_purch_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)
gen_purch_analysis = gen_purch_analysis[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]
gen_purch_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age = purchase_data["Age"].unique()
#age

bins = [0,10,14,19,24,29,34,39,49]
age_grps = ['< 10', '10 - 14', '15 - 19', '20 - 24', '25 - 29', '30 - 34', '35 - 39', '40+']


bin_df["Age Groups"] = pd.cut(purchase_data["Age"], bins, labels=age_grps)
bin_grp = bin_df.groupby(["Age Groups"])


bin_count = bin_grp["SN"].count()
tot_count = purchase_data["SN"].count()
percentage = (bin_count / tot_count) * 100
percentage


age_demographic = pd.DataFrame({"Total Count": bin_count,
                         "Percentage of Players": percentage})


age_demographic["Percentage of Players"] = age_demographic["Percentage of Players"].map("{:.2f}".format)
age_demographic
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age = purchase_data["Age"].unique()
#age

bins = [0,10,14,19,24,29,34,39,49]
age_grps = ['< 10', '10 - 14', '15 - 19', '20 - 24', '25 - 29', '30 - 34', '35 - 39', '40+']

age_bin_df = purchase_data.copy()
age_bin_df["Age Groups"] = pd.cut(purchase_data["Age"], bins, labels=age_grps)
#binColumn = pd.cut(purchase_data["Age"], bins, labels=age_grps)
bin_grp_df = age_bin_df.groupby(["Age Groups"])


purch_cnt = bin_grp_df["Age"].count()
avg_purch = bin_grp_df["Price"].mean()
tot_purch = bin_grp_df["Price"].sum()

new_bin_df = purchase_data.copy()
drop_dup = new_bin_df.drop_duplicates(subset='SN', keep="first")
new_bin_df["Age Groups"] = pd.cut(purchase_data["Age"], bins, labels=age_grps)
new_bin_df = new_bin_df.groupby(["Age Groups"])

avg_tot_purch_pp = (bin_grp_df["Price"].sum() / new_bin_df["SN"].count())
avg_tot_purch_pp


age_purch_analysis = pd.DataFrame({"Purchase Count": purch_cnt,
                         "Average Purchase Price": avg_purch,
                         "Total Purchase Value": tot_purch,
                         "Avg Total Purchase per Person": avg_tot_purch_pp})

age_purch_analysis["Average Purchase Price"] = age_purch_analysis["Average Purchase Price"].map("${:.2f}".format)
age_purch_analysis["Total Purchase Value"] = age_purch_analysis["Total Purchase Value"].map("${:.2f}".format)
age_purch_analysis["Avg Total Purchase per Person"] = age_purch_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)
age_purch_analysis = age_purch_analysis[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]
age_purch_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
grp_players = purchase_data.groupby(["SN"])
purch_count = grp_players["Item ID"].count()
tot_price = grp_players["Price"].sum()
avg_purch_price = (tot_price / purch_count)


top_spenders_df = pd.DataFrame({"Purchase Count": purch_count,
                         "Average Purchase Price": avg_purch_price,
                         "Total Purchase Value": tot_price})

top_spenders_df = top_spenders_df.sort_values("Total Purchase Value", ascending=False) 
top_spenders_df["Average Purchase Price"] = top_spenders_df["Average Purchase Price"].map("${:.2f}".format)
top_spenders_df["Total Purchase Value"] = top_spenders_df["Total Purchase Value"].map("${:.2f}".format)
top_spenders_df = top_spenders_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]
top_spenders_df.head(10)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items_df = purchase_data[["Item ID", "Item Name", "Price"]]
#items_df

grp_item = purchase_data.groupby(["Item ID", "Item Name"])
item_count = grp_item["SN"].count()
tot_price = grp_item["Price"].sum()
item_price = (tot_price / item_count)
tot_purch_value = (item_price * item_count)


popular_items_df = pd.DataFrame({"Purchase Count": item_count,
                          "Item Price": item_price,
                          "Total Purchase Value": tot_purch_value})

popular_items_df = popular_items_df.sort_values("Purchase Count", ascending=False) 
popular_items_df["Item Price"] = popular_items_df["Item Price"].map("${:.2f}".format)
popular_items_df["Total Purchase Value"] = popular_items_df["Total Purchase Value"].map("${:.2f}".format)
popular_items_df= popular_items_df[["Purchase Count", "Item Price", "Total Purchase Value"]]
popular_items_df.head(10)
```

```python
items_df = purchase_data[["Item ID", "Item Name", "Price"]]
#items_df

grp_item = purchase_data.groupby(["Item ID", "Item Name"])
item_count = grp_item["SN"].count()
tot_price = grp_item["Price"].sum()
item_price = (tot_price / item_count)
tot_purch_value = (item_price * item_count)


popular_items_df = pd.DataFrame({"Purchase Count": item_count,
                          "Item Price": item_price,
                          "Total Purchase Value": tot_purch_value})

popular_items_df = popular_items_df.sort_values("Total Purchase Value", ascending=False) 
popular_items_df["Item Price"] = popular_items_df["Item Price"].map("${:.2f}".format)
popular_items_df["Total Purchase Value"] = popular_items_df["Total Purchase Value"].map("${:.2f}".format)
popular_items_df= popular_items_df[["Purchase Count", "Item Price", "Total Purchase Value"]]
popular_items_df.head(10)
```
